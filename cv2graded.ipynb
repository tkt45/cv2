```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

def affine_transformations(image_path):
    # Load the image
    image = cv2.imread(image_path)

    # Define the transformation matrices
    rows, cols, _ = image.shape

    # Translation
    translation_matrix = np.float32([[1, 0, 50], [0, 1, 50]])
    translated_image = cv2.warpAffine(image, translation_matrix, (cols, rows))

    # Rotation
    rotation_matrix = cv2.getRotationMatrix2D((cols / 2, rows / 2), 45, 1)
    rotated_image = cv2.warpAffine(image, rotation_matrix, (cols, rows))

    # Scaling
    scaling_matrix = np.float32([[0.5, 0, 0], [0, 0.5, 0]])
    scaled_image = cv2.warpAffine(image, scaling_matrix, (cols, rows))

    # Shearing
    shearing_matrix = np.float32([[1, 0.5, 0], [0.5, 1, 0]])
    sheared_image = cv2.warpAffine(image, shearing_matrix, (cols, rows))

    return translated_image, rotated_image, scaled_image, sheared_image

# Example usage:
image_path = '/content/tushar1.jpeg'  # Upload your image to Colab and specify the path
transformed_images = affine_transformations(image_path)

# Display the original and transformed images
plt.figure(figsize=(12, 6))
plt.subplot(151), plt.imshow(cv2.cvtColor(cv2.imread(image_path), cv2.COLOR_BGR2RGB)), plt.title('Original')
plt.subplot(152), plt.imshow(cv2.cvtColor(transformed_images[0], cv2.COLOR_BGR2RGB)), plt.title('Translated')
plt.subplot(153), plt.imshow(cv2.cvtColor(transformed_images[1], cv2.COLOR_BGR2RGB)), plt.title('Rotated')
plt.subplot(154), plt.imshow(cv2.cvtColor(transformed_images[2], cv2.COLOR_BGR2RGB)), plt.title('Scaled')
plt.subplot(155), plt.imshow(cv2.cvtColor(transformed_images[3], cv2.COLOR_BGR2RGB)), plt.title('Sheared')

plt.show()
```
